```python
import pandas as pd

def csvtotext(input_csv_file, output_text_file, encoding='utf-8'):
    df = pd.read_csv(input_csv_file)
    df = df.head(100)

    with open(output_text_file, 'w', encoding=encoding) as text_file:
        for index, row in df.iterrows():
            text_file.write(str(row.iloc[0]) + '\n')

input_csv_file = 'quotes.csv'
output_text_file = 'extracted.txt'

csvtotext(input_csv_file, output_text_file)
```

```python
file_path = "extracted.txt"  # Update the file path if your file is in a different location

try:
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.read()
        print("File content successfully read and stored in a string.")

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
##assigning labels for each words as machine works on number
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
#creating the input output for the supervised model
input_seq=[]
for sentences in data.split('\n'):
  seq=tokenizer.texts_to_sequences([sentences])[0]

  for i in range(1,len(seq)):
    input_seq.append(seq[:i+1])
```

```python
input_seq
```

```python
##inorder to make array of equal lenmgth we do padding for it we require the max length of the sequence and rest will be padded with zero to make them all of equal length
max_len=max([len(x) for x in input_seq])
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_seq=pad_sequences(input_seq,maxlen=max_len,padding='pre')
```

```python
padded_input_seq
```

```python
##x-vale
x=padded_input_seq[:,:-1]
x
```

```python
##Yvalue
y=padded_input_seq[:,-1]
y
```

```python
##it is a multiclass classification task we will try to find the correct output
##by using softmax the one with highest probability will be the output for representation we will use one hot encoding
## example if output is 2 then [0,1,0,0,0.......]


x.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=1067)
```

```python
y.shape
```

```python
###now modeling part
##embeding->lstm->dense->output
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model=Sequential()
model.add(Embedding(1067,100,input_length=449))
model.add(LSTM(150))
model.add(Dense(1067,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=50)
```

```python
import numpy as np


text="dance like"
for i in range(15):
  ##tokenize
  token_text=tokenizer.texts_to_sequences([text])[0]
  ##padding
  padded_token_text=pad_sequences([token_text],maxlen=449,padding='pre')
  ##predict
  val=np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index==val:
      text=text + " " + word
      print(text)
```
